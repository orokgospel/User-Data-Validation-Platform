```python
#Importing the necessary libraries
import random
import datetime
import pandas as pd
```

```python
#Access codes for March 2023
#create a list container
code_db=[]
#Iterate random values with for loop
for ac in range(31):
    ac=random.randint(1000,9000)
    code_db.append(ac)
df= pd.DataFrame({'access_code': code_db})
#Create date for each code
df['date']=pd.date_range(start='3/01/2023', periods=len(df),freq='D')

#Export output to csv file
df.to_csv('access032023.csv')
```
